```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corona.csv')
```

```python
df
```

```python
df.groupby('infected').infected.count()
```

```python
target = df[['infected']]
target
```

```python
data = df.drop(['infected'], axis=1)
data
```

```python
data = data.to_numpy()
```

```python
data
```

```python
target = target.to_numpy()
```

```python
target
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

i = 1

for train_index, test_index in folds.split(data, target):
    
    if(i==1):
        train_data_1 = pd.DataFrame(data=data[train_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        test_data_1 = pd.DataFrame(data=data[test_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        train_target_1 = pd.DataFrame(data=target[train_index], columns=['infected'])
        test_target_1 = pd.DataFrame(data=target[test_index], columns=['infected'])
        train_1 = pd.concat([train_data_1, train_target_1], axis=1)
        test_1 = pd.concat([test_data_1, test_target_1], axis=1)
        
        train_1.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-1\train_1.csv', index = False, header=True)
        test_1.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-1\test_1.csv', index = False, header=True)
        
        
    if(i==2):
        train_data_2 = pd.DataFrame(data=data[train_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        test_data_2 = pd.DataFrame(data=data[test_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        train_target_2 = pd.DataFrame(data=target[train_index], columns=['infected'])
        test_target_2 = pd.DataFrame(data=target[test_index], columns=['infected'])
        train_2 = pd.concat([train_data_2, train_target_2], axis=1)
        test_2 = pd.concat([test_data_2, test_target_2], axis=1)
        train_2.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-2\train_2.csv', index = False, header=True)
        test_2.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-2\test_2.csv', index = False, header=True)
        
    if(i==3):
        train_data_3 = pd.DataFrame(data=data[train_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        test_data_3 = pd.DataFrame(data=data[test_index], columns=['fever','bodypain','age','runnynose','diffbreath'])
        train_target_3 = pd.DataFrame(data=target[train_index], columns=['infected'])
        test_target_3 = pd.DataFrame(data=target[test_index], columns=['infected'])
        train_3 = pd.concat([train_data_3, train_target_3], axis=1)
        test_3 = pd.concat([test_data_3, test_target_3], axis=1)
        train_3.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-3\train_3.csv', index = False, header=True)
        test_3.to_csv(r'C:\Users\Nafisa\Jupyter\covid19-symptoms-classification\Github Upload\With Ensemble Method\fold-3\test_3.csv', index = False, header=True)
    
    i = i+1
```
