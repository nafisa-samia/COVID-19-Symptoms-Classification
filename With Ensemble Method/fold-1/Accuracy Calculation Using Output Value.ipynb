```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import csv
```

```python
data1 = []
with open("output_value_1.csv") as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        data1.append(row)
        
data1
```

```python
data2 = []
with open("output_value_2.csv") as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        data2.append(row)
        
data2
```

```python
data3 = []
with open("output_value_3.csv") as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        data3.append(row)
        
data3
```

```python
data_Average = (np.array(data1) + np.array(data2) + np.array(data3)) / 3.0
```

```python
data_Average
```

```python
# Now Apply Winner take all to select class label from data_Average
class_label = np.argmax(data_Average, axis=1)
class_label
```

```python
test_data = pd.read_csv('test_1.csv')
test_data.sample(n=5)
```

```python
target_test = test_data[['infected']]
target_test.sample(n=5)
```

```python
targets = [[1,0],[0,1]]
test_array = np.array([targets[int(x)] for x in target_test.values[:,0:1]])
sample_test = np.argmax(test_array, axis=1)
```

```python
sample_test
```

```python
res = class_label == sample_test
res
```

```python
correct = np.sum(res)/len(res)
print('Ensemble Accuracy Using Average:',np.sum(res),'/',len(res), ':', (correct*100),'%')
```

```python
# Now Apply Winner take all to select class label from output data
data1_winner = np.argmax(data1, axis=1)
data1_winner_d = pd.DataFrame(data1_winner)

data2_winner = np.argmax(data2, axis=1)
data2_winner_d = pd.DataFrame(data2_winner)

data3_winner = np.argmax(data3, axis=1)
data3_winner_d = pd.DataFrame(data3_winner)

data_majority = pd.concat([data1_winner_d, data2_winner_d, data3_winner_d], axis=1)

data_majority
```

```python
data_majority['majority'] = data_majority.mode(axis=1)[0]
print(data_majority)
```

```python
majority = data_majority['majority'].to_numpy()
majority
```

```python
res_maj = majority == sample_test
res_maj
```

```python
correct_maj = np.sum(res_maj)/len(res_maj)
print('Ensemble Accuracy Using Majority Voting:',sum(res_maj),'/',len(res_maj), ':', (correct_maj*100),'%')
```

```python
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels

class_names = [0, 1]

def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = [0, 1]


    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plot_confusion_matrix(sample_test, class_label, classes=class_names, 
                      title='Confusion matrix for average')

# Plot normalized confusion matrix
plot_confusion_matrix(sample_test, majority, classes=class_names,
                      title='Confusion matrix for majority voting')

plt.show()
```
